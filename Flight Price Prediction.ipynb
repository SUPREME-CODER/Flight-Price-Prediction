```python
nrows, ncols = 6, 5
cities = clean_created_df['from'].unique()

fig, axes = mp.subplots(nrows=nrows, ncols=ncols, figsize=(20,30), sharey=True)
a_row, a_col = 0, 0
mp.subplots_adjust(wspace=0.1, hspace=1)
# fig.tight_layout()

for i in range(len(cities)):
    for j in range(len(cities)):
        if i == j:
            continue
        clean_ariline_group[cities[i]][cities[j]].plot(kind='bar', color='red', ax= axes[a_row][a_col])
        axes[a_row][a_col].set_title(cities[j], fontstyle='italic')
        axes[a_row][a_col].set_ylabel(cities[i], fontstyle='italic')
        a_col += 1
    a_row += 1
    a_col = 0
```
